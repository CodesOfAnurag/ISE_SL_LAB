# BlackFriday

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/BlackFriday.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df['City_Category'] = df['City_Category'].fillna("X")
df.head()
```

```python
df.City_Category = df.City_Category.map({'A' : 'Metro City', 'B' : 'Small Town', 'C' : 'Village', 'X':'Unknown'})
df.head()
```

```python
df1 = df[df.City_Category!='Unknown'].copy()
df1.Gender = df1.Gender.map({'F' : 'Female', 'M' : 'Male'})
```

```python
ax=sns.countplot(x='Gender', hue='City_Category', palette='Set1', data=df1)
ax.set(title='Gender with Count', xlabel='Gender', ylabel='Count')
plt.show()
```

# Titanic

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/titanictrain.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df.Embarked = df.Embarked.fillna("X")
df.head()
```

```python
df.Survived = df.Survived.map({0 : 'Dead', 1 : 'Alive'})
df.head()
```

```python
ax = sns.countplot(x='Pclass', hue='Survived', palette='Set3', data=df)
ax.set(title='Pclass with survival', xlabel='Class', ylabel='Count')
plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived'])
```
